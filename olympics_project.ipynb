```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import gradio as gr
```

```python
teams = pd.read_csv("/content/drive/MyDrive/Datasets /teams.csv")
teams = teams[['team', 'country', 'year', 'athletes', 'age', 'prev_medals', 'medals']]
teams = teams.dropna()
```

```python
# Split data into training and testing sets
train = teams[teams['year'] < 2012].copy()
test = teams[teams['year'] >= 2012].copy()
```

```python
# Define predictors and target
predictors = ['athletes', 'prev_medals']
target = 'medals'
```

```python
# Train the Linear Regression model
reg = LinearRegression()
reg.fit(train[predictors], train['medals'])
```

```python

# Define the prediction function
def predict_medals(athletes: int, prev_medals: int):
    """Predict the number of medals based on athletes and previous medals."""
    input_data = pd.DataFrame({'athletes': [athletes], 'prev_medals': [prev_medals]})
    prediction = reg.predict(input_data)[0]
    return max(0, round(prediction))  # Ensure no negative predictions
```

```python

# Gradio UI Setup (highlighted in blue)
interface = gr.Interface(  # 🟦 New Gradio interface setup
    fn=predict_medals,
    inputs=[
        gr.Number(label="Number of Athletes"),  # 🟦 Input field 1
        gr.Number(label="Previous Medals Won"),  # 🟦 Input field 2
    ],
    outputs="number",  # 🟦 Predicted output as a number
    title="Olympics Medal Prediction",  # 🟦 UI title
    description="Predict the number of medals a team might win based on the number of athletes and their previous medals."  # 🟦 UI description
)
```

```python

# Launch the interface
interface.launch(share='True')  # 🟦 Launch the UI
```
